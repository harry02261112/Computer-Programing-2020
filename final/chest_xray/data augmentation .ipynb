```python
%matplotlib inline 
import numpy as np 
import cv2
import matplotlib.pyplot as plt 
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    zca_whitening=False,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("COVID-19") if isfile(join("COVID-19", f))]
onlyfiles
```

```python
for i in range(len(onlyfiles)):
    img = cv2.imread("COVID-19/"+onlyfiles[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.reshape((1,) + img.shape)
    i = 0
    for batch in datagen.flow(img, batch_size=10,
                              save_to_dir='COVID-19', save_prefix='COVID-19-X-Ray', save_format='jpeg'):
        augImage = batch[0]
        augImage = augImage.astype('float32')
        augImage /= 255
        plt.imshow(augImage)
        i += 1
        if i > 19:
            break
```
